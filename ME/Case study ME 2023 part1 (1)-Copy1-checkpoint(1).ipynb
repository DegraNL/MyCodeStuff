*E. coli*  is een bekend micro-organisme dat glucose kan omzetten in acetaat, en daarna de gevormde acetaat weer kan consumeren. Het doel van deze case study is om te bepalen hoe er zoveel mogelijk biomassa gemaakt kan worden van een *E. coli* mutant (X<sub>1</sub>) die wel glucose kan omzetten in acetaat, maar de gevormde acetaat niet kan consumeren. De acetaatproductie is stoichiometrisch gekoppeld aan de glucose consumptie: voor iedere mol glucose geconsumeerd wordt er 0,5 mol acetaat geproduceerd [1–3].

**2b.**	Bereken, in Python, q<sub>glc</sub> op basis van de chemostaat data en je vergelijking verkregen in (2a).
Op basis van de gemeten C<sub>glc</sub> en de berekende q<sub>glc</sub> kunnen we q<sub>glcmaxX1</sub> and K<sub>glc</sub> schatten in Python.

Hiervoor gebruiken we lineaire regressie van de hyperbolische functie voor substraat opname:

$$
-q_{glc} = -q_{glc,X1}^{max}\frac{C_{glc}}{C_{glc}+K_{glc}}
$$

1.	Importeer de chemostaat data als een array in Python (je kunt data analyse library “pandas” hiervoor gebruiken) (1p)

5.	Gebruik de gedefinieerde functie, de gemeten Cglc¬, de berekende qglc and de curve_fit functie uit de scipy.optimize library, om qglcmaxX1 and Kglc te schatten. Print je gevonden waardes. (3p)
6.	Bereken de standaard deviaties van deze schattingen, waarbij sd = np.sqrt(np.diag(pcov)), zie https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html. Print je gevonden waardes. (3p)
7.  Maak een plot van C<sub>glc</sub> (x-as) vs -q<sub>glc</sub> (y-as) waarin jullie zowel de chemostaat data (als punten) als de curve op basis van jullie geschatte parameters (als lijn) laten zien.			(6p)




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

import pandas as pd
data = pd.read_excel( 'Parameter_estimation_data.xlsx', header = 0)  #data importeren vanuit excel voor de berekening van de data
data = np.asarray(data)

Fin= data[:,0]
Csin= data[:,1]
Cs= data[:,2]
Vl= data[:,3]
Cx= data[:,4]
Fph= data[:,5]

Fout= Fin + Fph                                                                                    #berekening outflow in 1/h
Mx=Vl*Cx                                                                                            #berekening hoeveelheid biomassa in de fermentor in Kg
mu= Fout/Vl                                                                                          #berekening van de groeisnelheid µ in h/L
qs= (Fout*Cs-Fin*Csin)/Mx                                                                 #berekening van de substraat opname Cmol glucose/Cmol biomassa/h

def MM(Cs, qsmax, Ks):                                     #een definition voor de bepaling van de concentratie substraat, de maximale substraat opname en de glucose affiniteitsconstante 
    return qsmax*Cs/(Ks+Cs)

popt, pcov = curve_fit(MM, Cs, qs)
qsmax, Ks = popt

sd = np.sqrt(np.diag(pcov))

#plotten van de gegevens 
Csmax = .1
Csplot = np.linspace(0,Csmax,100)
qsplot1 = -qsmax*Csplot/(Ks+Csplot)
plt.plot(Cs, -qs, 'ro', label= '-qs(t)')
plt.plot(Csplot, qsplot1, 'red')
plt.legend(loc='best')
plt.xlabel('Cglc (mol/kg)')
plt.ylabel('-qs (molS/CmolX/h)')
plt.grid()
plt.savefig('MM_fit.jpg', dpi=500)
plt.show()


print(f'Qsmax and Ks are {popt}')
print(sd)
```

# Vraag 3: Simuleer een batch cultuur van *E. coli* X1 in Python (18p)

**3a.** Construeer een python script waarmee je, op basis van je resultaten in vraag 1 en 2, een batch-cultuur van *E. coli* X1 kunt simuleren. Voer de simulatie uit en genereer plotjes van: (12p)					
1.	De concentraties van glucose, acetaat en biomassa tegen de tijd.
2.	q<sub>glc,X1</sub>, q<sub>ace,X1</sub> en μ<sub>X1</sub> tegen de tijd.
3.	R<sub>glc</sub>, R<sub>ace</sub> en R<sub>X1</sub> tegen de tijd.


```python
from scipy.integrate import odeint
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# model parameters
qsmax = -0.26                                       # Maximale specifieke glucose opname in Cmol glucose/Cmol biomass/h
Ks = 0.000235                                      # Glucose affiniteits constante in Cmol glucose/L
a = -0.91                                               # Herbert-Pirt parameter in Cmol glucose/Cmol biomass
Vl = 1                                                   # Volume van de fermentatie broth in Kg
time = 50                                              # totale stimulatie tijd in uur 



def batch(y, t):                                      #definition voor de balans vergelijkingen van substraat en biomassa 
    Cs, Cx, Cace = y 
    qs = qsmax*Cs/(Ks+Cs)
    ms = -0.01 -0.1 * Cace
    qace= -0.5*qs
    mu = (qs-ms)/a
    dCsdt = Cx*qs                                  #balans vergelijking voor de concentratie substraat(glucose)
    dCxdt = Cx*mu                                #balans vergelijking voor de concentratie biomassa 
    dCacedt = qace * Cx                         #balans vergelijking voor de concentratie acetaat 
    return [dCsdt, dCxdt, dCacedt]


#Integreren van de balansvergelijkingen
y0 = [0.2, 0.01, 0]
t = np.linspace(0, time, 1000)
sol = odeint(batch, y0, t)


Cs = sol[:, 0]
Cx = sol[:, 1]
Cace= sol[:,2]
ms= -0.01 - 0.1*Cace
qs = qsmax*Cs/(Ks+Cs)
qx = (qs-ms)/a
qace= -0.5*qs
Rs = qs*Cx*Vl
Rx = qx*Cx*Vl
Race = qace*Cx*Vl
Rc = -(Rs+Rx)                                        #vergelijking voor de productie van C atoom houdende producten  in deze batch 
Ro = (24*Rs+4.2*Rx+ 8*Race)/4          #vergelijking voor de productie van O atoom houdende producten in deze batch 
RQ = Rc/-Ro

#plotten van de gegeven 
plt.plot(t, Cs, 'b', label='Cs(t)')
plt.plot(t, Cx, 'g', label='Cx(t)')
plt.plot(t, Cace, 'r', label='Cace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Cs, Cx (Cmol/L)')
plt.grid()
plt.savefig('CsCx.jpg', dpi=500)
plt.show()

plt.plot(t, qs, 'r', label='qs(t)')
plt.plot(t, qx, 'c', label='qx(t)')
plt.plot(t, qace, 'g', label='qace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('qs, qx (Cmol/Cmol/h)')
plt.grid()
plt.savefig('qsqx.jpg', dpi=500)
plt.show()

plt.plot(t, Rs, 'm', label='Rs(t)')
plt.plot(t, Rx, 'y', label='Rx(t)')
plt.plot(t, Race, 'b', label='Race(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Rs, Rx (mol/h)')
plt.grid()
plt.savefig('RsRx.jpg', dpi=500)
plt.show()



```

De molecuulformule voor acetaat is C<sub>2</sub>H<sub>4</sub>O<sub>2</sub>. Onthoud dat in deze case study de glucose en acetaat concentraties gegeven zijn in mol, niet in Cmol.

**4f.**	Gebaseerd op je resultaten van vraag 4 (a) tot (e), stel, in Python, een model op voor de simulatie van een fed-batch cultuur. Let erop dat je fed-batch start nadat de glucose uit de batch-fase op is.
Produceer de volgende vier plots: (12p)
1.	De hoeveelheid glucose, acetaat en biomassa van *E. coli* X1 in de reactor tegen de tijd.

3.	R<sub>glc</sub>, R<sub>ace</sub> en R<sub>X1</sub> tegen de tijd.
4.	Het broth gewicht tegen de tijd.


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#model parameters 
qsmax = -0.26                                       # Maximale specifieke glucose opmane in Cmol glucose/Cmol biomass/h
Ks = 0.000235                                      # Glucose affiniteits constante in Cmol glucose/L
a = -0.91                                               # Herbert-Pirt parameter in Cmol glucose/Cmol biomass
feed = 0.005                                         # feed rate van de fed-batch in kg/uur
Csin = 2.5                                            # glucose concentration in the feed Cmol/kg 
tbatch = 11.2                                           #tijd dat de batch fase verloopt in uur 
time = 100                                             # totale stimulatie tijd in uur (batch + fed-batch)



def fedbatch(y, t):                                 #model voor de substraat en biomassa concentraties 
    Ms, Mx, Mb, Mace = y
    if t < tbatch:
        inflow = 0
    else:
        inflow = feed
    ms = -0.01 -0.1 * (Mace/Mb)
    Cs = Ms/Mb
    qs = qsmax*Cs/(Ks+Cs)
    qace= -0.5*qs
    mu = (qs-ms)/a
    
    Rc = -(6*qs*Mx + mu*Mx + 2*qace*Mx)
    Ro = (24*qs*Mx + 4.2*mu*Mx + 8*qace*Mx)/4
    dMsdt = Mx*qs + inflow*Csin           #balans vergelijking voor de hoeveelheid substraat in de fermentor 
    dMxdt = (mu)*Mx                             #balans vergelijking voor de hoeveelheid biomassa in de fermentor 
    dMacedt= qace*Mx                           #balans vergelijking voor de hoeveelhied acetaat in de fermentor 
    dMbdt = inflow -0.044*Rc - 0.032*Ro                    #balans vergelijking voor de broth weight 
    
    
    return [dMsdt, dMxdt, dMbdt, dMacedt]


#integreren van de balansvergelijkingen
y0 = [0.2, 0.01, 1, 0]
t = np.linspace(0, time, 1000)
sol = odeint(fedbatch, y0, t)

Ms = sol[:, 0]
Mx = sol[:, 1]
Mb = sol[:, 2]
Mace=sol[:, 3]

Cs = Ms/Mb
Cx = Mx/Mb

qs = qsmax*Cs/(Ks+Cs)
qace= -0.5*qs
mu = (qs-ms)/a

Rs = qs*Mx
Rx = mu*Mx
Race=  qace*Mx
Rc = -( 6*Rs+ Rx + 2*Race )                                #vergelijking voor de productie van C atoom houdende producten  in deze batch
Ro = (24*Rs + 4.2*Rx + 8*Race)/4          #vergelijking voor de productie van O atoom houdende producten in deze batch 

# plotten van de gegeven 
plt.plot(t, Ms, 'b', label='Ms(t)')
plt.plot(t, Mx, 'g', label='Mx(t)')
plt.plot(t, Mace, 'r', label='Mace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Ms, Mxv, (Cmol)')
plt.grid()
plt.savefig('FB2_MsMx.jpg', dpi=500)
plt.show()

plt.plot(t, qs, 'r', label='qs(t)')
plt.plot(t, qace, 'b', label='qace(t)')
plt.plot(t, mu, 'c', label='mu(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('qs, mu (Cmol/Cmol/h)')
plt.grid()
plt.savefig('FB2_qsqx.jpg', dpi=500)
plt.show()

plt.plot(t, Rs, 'm', label='Rs(t)')
plt.plot(t, Rx, 'y', label='Rx(t)')
plt.plot(t, Race, 'b', label='Race(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Rs, Rx (Cmol/h)')
plt.grid()
plt.savefig('FB2_RsRx.jpg', dpi=500)
plt.show()

plt.plot(t, Mb, 'brown', label='Mb(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Mb (kg)')
plt.grid()
plt.savefig('FB2_Mb.jpg', dpi=500)
plt.show()

```

**5e.**	Gebaseerd op je resultaten van vraag 5 (a) tot (d), stel, in Python, een model op voor de simulatie van een fed-batch co-cultuur van *E. coli X1* en *E. coli X2*. Let erop dat je fed-batch start direct nadat de glucose uit de batch-fase op is.

Produceer de volgende vier plots van je simulatie: 					(12p)
1.	De totale hoeveelheden glucose, acetaat en biomassa van *E. coli X1* en *E. coli X2* tegen de tijd. 
2.	q<sub>glc,X1</sub>, q<sub>ace,X1</sub>, q<sub>ace,X2</sub>, μ<sub>X1</sub> en μ<sub>X2</sub> tegen de tijd. 
3.	R<sub>glc</sub>, R<sub>ace,X1</sub>, R<sub>ace,X2</sub>, R<sub>ace,totaal</sub>, R<sub>X1</sub> en R<sub>X2</sub> tegen de tijd. 
4.  m<sub>s</sub> tegen de tijd.
4.	Het broth gewicht tegen de tijd.


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#model parameters 
qsmax_x1 = -0.26                                            # Maximale specifieke glucose opmane in Cmol glucose/Cmol biomass/h
Ks_x1 = 0.000235                                            # Glucose affiniteits constante in Cmol glucose/L
a_x1 = -0.91                                                # Herbert-Pirt parameter in Cmol glucose/Cmol biomass
Ks_x2 = 0.0017                                           # glucose affinity constant (mol acetaat/Kg)
a_x2 = -1.17                                             # Herbert-Pirt parameter (mol acetaat/Cmol biomass)
ms_x2 = -0.005                                           # Maintenance X2 (mol acetaat/CmolX2)
qsmax_x2 = -0.06                                         # max specific glucose uptake rate (mol acetaat/Cmol biomass/h)

feed = 0.005                                             # feed rate van de fed-batch in kg/uur
Csin = 2.5                                               # glucose concentration in the feed Cmol/kg 
tbatch = 11.2                                            #tijd dat de batch fase verloopt in uur 
time = 500                                               # totale stimulatie tijd in uur (batch + fed-batch)


def fedbatch(y, t):                                 #model voor de substraat en biomassa concentraties 
    Ms, M_x1, M_x2, Mb, Mace = y
    if t < tbatch:
        inflow = 0
    else:
        inflow = feed

    ms_x1 = -0.01 -0.1 * (Mace/Mb)
    Cglc = Ms/Mb
    Cace= Mace/Mb

    qglc_x1 = qsmax_x1*Cglc/(Ks_x1+Cglc)
    qace_x2 = qsmax_x2*Cace/(Ks_x2+Cace)
    qace_x1= -0.5*qglc_x1
    
    mu_x1 = (qglc_x1-ms_x1)/a_x1
    mu_x2 = (qace_x2-ms_x2)/a_x2

    
    Rc = -(6*qglc_x1*M_x1 + mu_x1*M_x1  + mu_x2* M_x2  + 2*qace_x1*M_x1 + 2*qace_x2 *M_x2)
    Ro = (24*qglc_x1*M_x1 + 4.2*mu_x1*M_x1 + 4.2*mu_x2*M_x2 + 8*qace_x1*M_x1 + 8*qace_x2 * M_x2)/4

    dMglcdt = M_x1*qglc_x1 + inflow*Csin           #balans vergelijking voor de hoeveelheid substraat in de fermentor 
    dMx1dt = (mu_x1)*M_x1                          #balans vergelijking voor de hoeveelheid biomassa in de fermentor 
    dMx2dt = (mu_x2)*M_x2  
    dMacedt= qace_x1*M_x1 + qace_x2 * M_x2                         #balans vergelijking voor de hoeveelhied acetaat in de fermentor 
    dMbdt = inflow -0.044*Rc - 0.032*Ro                    #balans vergelijking voor de broth weight 
    
    
    return [dMglcdt, dMx1dt, dMx2dt, dMbdt, dMacedt]


#integreren van de balansvergelijkingen
y0 = [0.2, 0.01, 0.002, 1, 0]
t = np.linspace(0, time, 1000)
sol = odeint(fedbatch, y0, t)

Mglc = sol[:, 0]
M_x1 = sol[:, 1]
M_x2 = sol[:, 2]
Mb=sol[:, 3]
Mace=sol[:, 4]

Cglc = Mglc/Mb
Cx1 = M_x1/Mb
Cace= Mace/Mb
ms_x1 = -0.01 -0.1 * (Mace/Mb)

qace_x2 = qsmax_x2*Cace/(Ks_x2+Cace)
qglc_x1 = qsmax_x1*Cglc/(Ks_x1+Cglc)
qace_x1= -0.5*qglc_x1

mu_x1 = (qglc_x1-ms_x1)/a_x1
mu_x2 = (qace_x2-ms_x2)/a_x2

Rglc = qglc_x1*M_x1
Rx1 = mu_x1*M_x1
Rx2 = mu_x2*M_x2
Race_x1=  qace_x1*M_x1 
Race_x2 = qace_x2 * M_x2
Race= Race_x1 + Race_x2
Rc = -(6*qglc_x1*M_x1 + mu_x1*M_x1  + mu_x2* M_x2  + 2*qace_x1*M_x1 + 2*qace_x2 *M_x2)
Ro = (24*qglc_x1*M_x1 + 4.2*mu_x1*M_x1 + 4.2*mu_x2*M_x2 + 8*qace_x1*M_x1 + 8*qace_x2 * M_x2)/4

# plotten van de gegeven 
plt.plot(t, Mglc, 'b', label='Ms(t)')
plt.plot(t, M_x1, 'g', label='Mx1(t)')
plt.plot(t, M_x2, 'y', label='Mx2(t)')
plt.plot(t, Mace, 'r', label='Mace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Ms, Mxv, (Cmol)')
plt.grid()
plt.savefig('FB2_MsMx.jpg', dpi=500)
plt.show()

plt.plot(t, qglc_x1, 'r', label='qs(t)')
plt.plot(t, qace_x1, 'b', label='qace x1(t)')
plt.plot(t, qace_x2, 'g', label='qace x2(t)')
plt.plot(t, mu_x1, 'c', label='mu x1(t)')
plt.plot(t, mu_x2, 'y', label='mu x2(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('qs, mu (Cmol/Cmol/h)')
plt.grid()
plt.savefig('FB2_qsqx.jpg', dpi=500)
plt.show()

plt.plot(t, Rglc, 'm', label='Rglc(t)')
plt.plot(t, Rx1, 'y', label='Rx1(t)')
plt.plot(t, Rx2, 'r', label='Rx2(t)')
plt.plot(t, Race_x1, 'b', label='RaceX1(t)')
plt.plot(t, Race_x1, 'brown', label='RaceX2(t)')
plt.plot(t, Race, 'green', label='Race(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Rs, Rx (Cmol/h)')
plt.grid()
plt.savefig('FB2_RsRx.jpg', dpi=500)
plt.show()

plt.plot(t, ms_x1, 'brown', label='ms_x1(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Ms (kg)')
plt.grid()
plt.savefig('FB2_Mb.jpg', dpi=500)
plt.show()

plt.plot(t, Mb, 'brown', label='Mb(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Mb (kg)')
plt.grid()
plt.savefig('FB2_Mb.jpg', dpi=500)
plt.show()

```
