# Case study ME 2023 deel 1 Biomassa productie door *Escherichia coli* (144p)

*E. coli*  is een bekend micro-organisme dat glucose kan omzetten in acetaat, en daarna de gevormde acetaat weer kan consumeren. Het doel van deze case study is om te bepalen hoe er zoveel mogelijk biomassa gemaakt kan worden van een *E. coli* mutant (X<sub>1</sub>) die wel glucose kan omzetten in acetaat, maar de gevormde acetaat niet kan consumeren. Voor dit micro-organisme is bekend dat de maintenance coëfficient afhankelijk is van de acetaat concentratie: m<sub>glc,x1</sub> = -0,01 – 0,1∙C<sub>ace</sub> mol glucose / Cmol X<sub>1</sub> / h. Acetaat is hier het katabole product, dus er is geen qp om rekening mee te houden. De acetaatproductie is stoichiometrisch gekoppeld aan de glucose consumptie: voor iedere mol glucose geconsumeerd wordt er 0,5 mol acetaat geproduceerd [1–3].

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

import pandas as pd
data = pd.read_excel( 'Parameter_estimation_data.xlsx', header = 0)
data = np.asarray(data)

Fin= data[:,0]
Csin= data[:,1]
Cs= data[:,2]
Vl= data[:,3]
Cx= data[:,4]
Fph= data[:,5]

Fout= Fin + Fph
Mx=Vl*Cx
mu= Fout/Vl
qs= (Fout*Cs-Fin*Csin)/Mx

def MM(Cs, qsmax, Ks):
    return qsmax*Cs/(Ks+Cs)

popt, pcov = curve_fit(MM, Cs, qs)
qsmax, Ks = popt

sd = np.sqrt(np.diag(pcov))

Csmax = .1
Csplot = np.linspace(0,Csmax,100)
qsplot1 = -qsmax*Csplot/(Ks+Csplot)
plt.plot(Cs, -qs, 'ro', label= '-qs(t)')
plt.plot(Csplot, qsplot1, 'red')
plt.legend(loc='best')
plt.xlabel('Cglc (mol/kg)')
plt.ylabel('-qs (molS/CmolX/h)')
plt.grid()
plt.savefig('MM_fit.jpg', dpi=500)
plt.show()


print(f'Qsmax and Ks are {popt}')
print(sd)
```

# Vraag 3: Simuleer een batch cultuur van *E. coli* X1 in Python (18p)

**3a.** Construeer een python script waarmee je, op basis van je resultaten in vraag 1 en 2, een batch-cultuur van *E. coli* X1 kunt simuleren. Voer de simulatie uit en genereer plotjes van: (12p)					
1.	De concentraties van glucose, acetaat en biomassa tegen de tijd.
2.	q<sub>glc,X1</sub>, q<sub>ace,X1</sub> en μ<sub>X1</sub> tegen de tijd.
3.	R<sub>glc</sub>, R<sub>ace</sub> en R<sub>X1</sub> tegen de tijd.


```python
from scipy.integrate import odeint
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# model parameters
qsmax = -0.26  # max specific glucose uptake rate (Cmol glucose/Cmol biomass/h)
Ks = 0.000235     # glucose affinity constant (Cmol glucose/L)
a = -0.91     # Herbert-Pirt parameter (Cmol glucose/Cmol biomass)
Vl = 1        # broth volume (kg)
time = 50     # total simulation time (hours)


# model definition (balance equations for substrate and biomass)
def batch(y, t):
    Cs, Cx, Cace = y
    qs = qsmax*Cs/(Ks+Cs)
    ms = -0.01 -0.1 * Cace
    qace= -0.5*qs
    mu = (qs-ms)/a
    dCsdt = Cx*qs
    dCxdt = Cx*mu
    dCacedt = qace * Cx
    return [dCsdt, dCxdt, dCacedt]


# Integration
y0 = [0.2, 0.01, 0]
t = np.linspace(0, time, 1000)
sol = odeint(batch, y0, t)


Cs = sol[:, 0]
Cx = sol[:, 1]
Cace= sol[:,2]
ms= -0.01 - 0.1*Cace
qs = qsmax*Cs/(Ks+Cs)
qx = (qs-ms)/a
qace= -0.5*qs
Rs = qs*Cx*Vl
Rx = qx*Cx*Vl
Race = qace*Cx*Vl
Rc = -(Rs+Rx)
Ro = (24*Rs+4.2*Rx+ 8*Race)/4
RQ = Rc/-Ro

# plotting
plt.plot(t, Cs, 'b', label='Cs(t)')
plt.plot(t, Cx, 'g', label='Cx(t)')
plt.plot(t, Cace, 'r', label='Cace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Cs, Cx (Cmol/L)')
plt.grid()
plt.savefig('CsCx.jpg', dpi=500)
plt.show()

plt.plot(t, qs, 'r', label='qs(t)')
plt.plot(t, qx, 'c', label='qx(t)')
plt.plot(t, qace, 'g', label='qace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('qs, qx (Cmol/Cmol/h)')
plt.grid()
plt.savefig('qsqx.jpg', dpi=500)
plt.show()

plt.plot(t, Rs, 'm', label='Rs(t)')
plt.plot(t, Rx, 'y', label='Rx(t)')
plt.plot(t, Race, 'b', label='Race(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Rs, Rx (mol/h)')
plt.grid()
plt.savefig('RsRx.jpg', dpi=500)
plt.show()



```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# model parameters
qsmax = -0.26    # max specific glucose uptake rate (Cmol glucose/Cmol biomass/h)
Ks = 0.000235     # glucose affinity constant (Cmol glucose/kg)
a = -0.91      # Herbert-Pirt parameter (Cmol glucose/Cmol biomass)
feed = 0.005   # feed rate of the fed-batch (kg/h)
Csin = 2.5   # glucose concentration in the feed (Cmol/kg)
tbatch = 11  # duration of the batch phase (hours)
time = 100   # total runtime simulation time (hours)


# model definition (balance equations for substrate and biomass)
def fedbatch(y, t):
    Ms, Mx, Mb, Cace, Mace = y
    if t < tbatch:
        inflow = 0
    else:
        inflow = feed
    ms = -0.01 -0.1 * Cace
    Cs = Ms/Mb
    qs = qsmax*Cs/(Ks+Cs)
    qace= -0.5*qs
    mu = (qs-ms)/a
    
    Rc = -(qs*Mx + mu*Mx +qace*Mx)
    Ro = (24*qs*Mx + 4.2*mu*Mx+ 8*qace*Mx)/4
    dMsdt = Mx*qs+inflow*Csin
    dMxdt = (mu)*Mx
    dMacedt= qace*Mx
    dMbdt = inflow -Rc- Ro
    
    
    return [dMsdt, dMxdt, dMbdt, Cace, dMacedt]


# Integration
y0 = [0.2, 0.01, 1, 0, 0]
t = np.linspace(0, time, 1000)
sol = odeint(fedbatch, y0, t)

Ms = sol[:, 0]
Mx = sol[:, 1]
Mb = sol[:, 2]
Mace=sol[:,4]
Cs = Ms/Mb
Cx = Mx/Mb
qs = qsmax*Cs/(Ks+Cs)
mu = (qs-ms)/a
Rs = qs*Mx
Rx = mu*Mx
Race=  qace*Mx
Rc = -(Rs+Rx+ Race)
Ro = (24*Rs + 4.2*Rx + 8*Race)/4
RQ = Rc/-Ro

# plotting
plt.plot(t, Ms, 'b', label='Ms(t)')
plt.plot(t, Mx, 'g', label='Mx(t)')
plt.plot(t, Mace, 'r', label='Mace(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Ms, Mxv, (Cmol)')
plt.grid()
plt.savefig('FB2_MsMx.jpg', dpi=500)
plt.show()

plt.plot(t, qs, 'r', label='qs(t)')
plt.plot(t, mu, 'c', label='mu(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('qs, mu (Cmol/Cmol/h)')
plt.grid()
plt.savefig('FB2_qsqx.jpg', dpi=500)
plt.show()

plt.plot(t, mu, 'c', label='mu(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('mu (Cmol/Cmol/h)')
plt.axis([0, 100, 0, 0.3])
plt.grid()
plt.savefig('FB2_mu_zoom.jpg', dpi=500)
plt.show()

plt.plot(t, Rs, 'm', label='Rs(t)')
plt.plot(t, Rx, 'y', label='Rx(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Rs, Rx (Cmol/h)')
plt.grid()
plt.savefig('FB2_RsRx.jpg', dpi=500)
plt.show()

plt.plot(t, Ro, 'purple', label='Ro(t)')
plt.plot(t, Rc, 'grey', label='Rc(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Ro, Rc (mol/h)')
plt.grid()
plt.savefig('FB2_RoRc.jpg', dpi=500)
plt.show()

plt.plot(t, RQ, 'red', label='RQ(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('RQ (-)')
plt.grid()
plt.savefig('FB2_RQ.jpg', dpi=500)
plt.show()

plt.plot(t, Mb, 'brown', label='Mb(t)')
plt.legend(loc='best')
plt.xlabel('t (hours)')
plt.ylabel('Mb (kg)')
plt.grid()
plt.savefig('FB2_Mb.jpg', dpi=500)
plt.show()

```
